```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/dating.csv'

df = pd.read_csv(file_url)
```

```python
df = df.dropna(subset = ['pref_o_attractive', 'pref_o_sincere','pref_o_intelligence','pref_o_funny','pref_o_ambitious','pref_o_shared_interests','attractive_important','sincere_important','intellicence_important','funny_important','ambtition_important','shared_interests_important'])
```

```python
df = df.fillna(-99)
```

```python
def age_gap(x):
    if x['age'] == -99:
        return -99
    elif x['age_o'] == -99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age']
    else:
        return x['age'] - x['age_o']
    
```

```python
df['age_gap'] = df.apply(age_gap, axis=1)
```

```python
df['age_gap_abs'] = abs(df['age_gap'])
```

```python
def same_race(x):
    if x['race'] == -99:
        return -99
    elif x['race_o'] == -99:
        return -99
    elif x['race'] == x['race_o']:
        return 1
    else:
        return -1
```

```python
df['same_race'] = df.apply(same_race, axis=1)
```

```python
def rating(df, importance, score):
    if df[importance] == -99:
        return -99
    elif df[score] == -99:
        return -99
    else:
        return df[importance]*df[score]
```

```python
df.columns[8:14]
```

```python
partner_imp = df.columns[8:14]
partner_rate_me = df.columns[14:20]
my_imp = df.columns[20:26]
my_rate_partner = df.columns[26:32]
```

```python
new_label_partner = ['attractive_p','sincere_partner_p','intelligence_p','funny_p','ambition_p','shared_interests_p']
new_label_me = ['attractive_m','sincere_partner_m','intelligence_m','funny_m','ambition_m','shared_interests_m']
```

```python
for i,j,k in zip(new_label_partner, partner_imp, partner_rate_me):
    df[i] = df.apply(lambda x: rating(x,j,k), axis=1)
```

```python
for i,j,k in zip(new_label_me, my_imp, my_rate_partner):
    df[i] = df.apply(lambda x: rating(x, j, k), axis=1)
```

```python
df = pd.get_dummies(df, columns = ['gender','race','race_o'], drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('match',axis=1), df['match'], test_size =0.2, random_state =100)
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier(n_estimators = 500, max_depth = 5, random_state=100)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_test, pred)
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```
